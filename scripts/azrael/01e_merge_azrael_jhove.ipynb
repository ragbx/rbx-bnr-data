```python
import pandas as pd
import os

all_results = []
files = [f for f in os.listdir('data/jhove_csv')]
for filename in sorted(files):
    print(filename)
    tmp = pd.read_csv(f"data/jhove_csv/{filename}", usecols=['jhove_uri', 'jhove_lastModified', 'jhove_format', 'jhove_version',
       'jhove_mimeType', 'jhove_size', 'jhove_imageWidth', 'jhove_imageHeight',
       'jhove_samplingFrequencyUnit', 'jhove_x_resolution',
       'jhove_y_resolution'])
    all_results.append(tmp)
```

```python
len(all_results)
```

```python
all_results[0].columns
```

```python
df = pd.DataFrame(columns=['jhove_uri', 'jhove_lastModified', 'jhove_format', 'jhove_version',
       'jhove_mimeType', 'jhove_size', 'jhove_imageWidth', 'jhove_imageHeight',
       'jhove_samplingFrequencyUnit', 'jhove_x_resolution',
       'jhove_y_resolution'])
df
```

```python
df = pd.concat(all_results)
```

```python
df
```

```python
df.to_csv("data/jhove_synthese_20240615.csv.gz", index=False)
```

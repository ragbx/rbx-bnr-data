```python
import pandas as pd
from bnr.azrael import convert_size
```

```python
df = pd.read_csv("data/bnr_azrael_20240611_detailed.csv.gz")
```

```python
df['nb'] = 1
```

```python
df.columns
```

```python
df = df[df['path0'].str[0:3] == 'BNR']
```

```python
df_pivot = pd.pivot_table(df, values=['nb', 'size'], index=['path0'],
                       columns=['file_type'], aggfunc='sum', fill_value=0, margins=True, margins_name='Total')
df_pivot
```

```python
main_gr['size_go'] = df_pivot['size'].apply(lambda x: convert_size(x, to_size='go'))
```

```python
main_gr = df.groupby(['path0', 'file_type']).agg(
        {
            'nb': 'sum',
            'size': 'sum'
        }
    ).reset_index()
main_gr['size_go'] = main_gr['size'].apply(lambda x: convert_size(x, to_size='go'))
df_pivot = pd.pivot_table(main_gr, values=['nb', 'size_go'], index=['path0'],
                       columns=['file_type'], aggfunc='sum', fill_value=0, margins=True, margins_name='Total')
df_pivot.to_excel("test.xlsx")
```
